#Implementación de los algoritmos, J4.5, C50 de los árboles de decisión y el algoritmo KNN

En este proyecto, utilizaremos el dataset iris para implementar tres algoritmos de clasificación: J4.5, C50, y KNN. Cada algoritmo tiene sus propias características que permiten explorar diferentes formas de clasificar las flores iris según sus características (longitud y ancho de sépalos y pétalos).
Objetivos






##1. Implementación del Algoritmo J4.5

El algoritmo J4.5 es una variante del Árbol de Decisión que utiliza el criterio de entropía para medir la cantidad de información obtenida en cada división de los datos. Es ideal para problemas de clasificación con datos categóricos o numéricos. En sklearn, simulamos este comportamiento usando DecisionTreeClassifier con el argumento criterion='entropy'.


```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt

# Cargar el dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo J4.5 con entropía
j45 = DecisionTreeClassifier(criterion='entropy', random_state=42)
j45.fit(X_train, y_train)

# Evaluar el modelo
y_pred_j45 = j45.predict(X_test)
print("Precisión del modelo J4.5:", accuracy_score(y_test, y_pred_j45))
print("\nReporte de clasificación J4.5:\n", classification_report(y_test, y_pred_j45))

# Visualizar el Árbol de Decisión
plt.figure(figsize=(8, 12))
plot_tree(j45, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.title("Árbol de Decisión J4.5")
plt.show()
```

1. Si el ancho del pétalo (petal width) <= 0.6:
   Clase: Iris-setosa

2. Si el ancho del pétalo (petal width) > 0.6:
   2.1. Si el ancho del pétalo (petal width) <= 1.7:
        - Si la longitud del pétalo (petal length) <= 4.9:
          Clase: Iris-versicolor
        - Si la longitud del pétalo (petal length) > 4.9:
            - Si el ancho del pétalo (petal width) <= 1.5:
              Clase: Iris-virginica
            - Si el ancho del pétalo (petal width) > 1.5:
              Clase: Iris-versicolor

   2.2. Si el ancho del pétalo (petal width) > 1.7:
        Clase: Iris-virginica


#2. Implementación del Algoritmo C50

El algoritmo C50 es una mejora de J4.5 que optimiza la clasificación utilizando Árboles de Decisión. En Python, no está directamente disponible, pero podemos aproximar su comportamiento utilizando DecisionTreeClassifier con el criterio gini.


```python
c50 = DecisionTreeClassifier(criterion='gini', random_state=42)
c50.fit(X_train, y_train)

# Evaluar el modelo
y_pred_c50 = c50.predict(X_test)
print("Precisión del modelo C50:", accuracy_score(y_test, y_pred_c50))
print("\nReporte de clasificación C50:\n", classification_report(y_test, y_pred_c50))

# Visualizar el Árbol de Decisión
plt.figure(figsize=(12, 8))
plot_tree(c50, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.title("Árbol de Decisión C50")
plt.show()
```

El resultado del algoritmo C50 fue bastante preciso, mostrando los siguentes resultados en el ábol

1. Si el ancho del pétalo (petal width) <= 0.6:
   Clase: Iris-setosa

2. Si el ancho del pétalo (petal width) > 0.6:
   2.1. Si el ancho del pétalo (petal width) <= 1.7:
        - Si la longitud del pétalo (petal length) <= 4.9:
          Clase: Iris-versicolor
        - Si la longitud del pétalo (petal length) > 4.9:
            - Si el ancho del pétalo (petal width) <= 1.5:
              Clase: Iris-virginica
            - Si el ancho del pétalo (petal width) > 1.5:
              Clase: Iris-versicolor

   2.2. Si el ancho del pétalo (petal width) > 1.7:
        Clase: Iris-virginica

#3. Implementación del Algoritmo KNN

El algoritmo KNN clasifica los datos según la proximidad entre puntos utilizando la distancia Euclidiana. Es un método no paramétrico que funciona especialmente bien con datos etiquetados.

```python
from sklearn.neighbors import KNeighborsClassifier

# Crear y entrenar el modelo KNN con k=3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Evaluar el modelo
y_pred_knn = knn.predict(X_test)
print("Precisión del modelo KNN:", accuracy_score(y_test, y_pred_knn))
print("\nReporte de clasificación KNN:\n", classification_report(y_test, y_pred_knn))
```

Los resultados del algoritmo KNN aplicado al dataset iris muestran un excelente desempeño, logrando una precisión del 100% en el conjunto de prueba. Esto significa que el modelo fue capaz de clasificar correctamente cada flor en su respectiva clase (Iris-setosa, Iris-versicolor o Iris-virginica). El uso del valor k=3 permitió tomar en cuenta los tres vecinos más cercanos para cada predicción, lo que fue ideal para este dataset bien estructurado y con características claramente diferenciadas.
