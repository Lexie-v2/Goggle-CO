```python
import matplotlib.pyplot as plt

#Verificar que el número es primo
def primo(n):
  if n <= 1:
    return False
  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False
  return True

#Verificar que el número es par
def par(n):
  return n % 2 == 0

#Verificar que el número es perfecto
def perfecto(n):
  suma = 0
  for i in range(1, n):
    if n % i == 0:
      suma += i
  return suma == n

def contar_numeros(rango):
  primos = sum(1 for n in rango if primo(n))
  pares = sum(1 for n in rango if par(n))
  perfectos = sum(1 for n in rango if perfecto(n))
  return primos, pares, perfectos

def graficar(primos, pares, perfectos):
  etiquetas = ['Primos', 'Pares', 'Perfectos']
  valores = [primos, pares, perfectos]
  colores = ['gold', 'yellowgreen', 'lightcoral']

  plt.figure(figsize=(8, 6))
  plt.pie(valores, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=140)
  plt.axis('equal')
  plt.title('Distribución de Números')
  plt.show()

def main():
  n1 = int(input("Ingresa el primer número: "))
  n2 = int(input("Ingresa el segundo número: "))

  menor, mayor = min(n1, n2), max(n1, n2)
  rango = range(menor, mayor + 1)

  primos, pares, perfectos = contar_numeros(rango)

  print("Números primos:", primos)
  print("Números pares:", pares)
  print("Números perfectos:", perfectos)

  graficar(primos, pares, perfectos)

main()
```
